```python
# Complete
# Select day of analysis
month = 12
year = 2021
days = [str(i).zfill(2) for i in range(1, monthrange(year, month)[1] + 1)]
#days = [str(i).zfill(2) for i in range(1, 2 + 1)]
month_date = date(year,month, 1)
rows = ['MEDIA [mV]', 'STD [%]', 'MIN [mV]','MAX [mV]', 'MEDIANA [mV]', 'MODA [mV]']
if machine == "windows":
        folder_images = mtu_per_days + month_date.strftime("%b_%Y").upper()

if not os.path.exists(folder_images):
        os.makedirs(folder_images)

for day in days:

    day = int(day)
    day_date = date(year,month, day)

    # Load mtu file

    dir_mtu =  mtu_file_name + day_date.strftime("%b%d_%Y").upper() + ".txt"
    mtu = pd.read_csv(dir_mtu, delimiter = "\t")

    # Normalize data
    sensors = ["P" + str(i) for i in range(1, 73)]
    sensors0 = ["P" + str(i) for i in[15, 16, 17, 19, 21, 22, 23, 50, 51, 52, 58, 59, 60]]

    for sensor in sensors0:
        sensors.remove(sensor)

    # Here goes the plot
    mydpi = 200
    ancho = 1920
    ratio = 10/4
    movil = 30
    colors = ['#FF0000','#FF4000','#FF8000','#FFC000','#ffff00','#bfff00','#00ff00','#00ff80','#00ffff','#00bfff','#007fff','#0000ff','#8000ff','#c000ff','#ff00ff','#ff007f']
    listas = [sensors[i:i + 16] for i in range(0, len(sensors), 16)]

    if machine == "windows":
        folder_images_child = day_date.strftime("%b%d_%Y").upper()

    if not os.path.exists(folder_images + tab + folder_images_child):
        os.makedirs(folder_images + tab + folder_images_child)

    mtu = mtu[sensors]

    # Normalize sensors
    for sensor in sensors:
        max = mtu[sensor].max()

        if max != 0:
            mtu[sensor] = mtu[sensor]/max
        else:
            mtu[sensor] = 0


    j = 0
    grupos = np.arange(1, len(listas) + 1)
    for lista in listas:
        fig, ax = plt.subplots(1, figsize= ( ancho/mydpi , (ancho/ratio)/mydpi ) , dpi = mydpi)
        plt.rcParams.update({'font.family':'Courier New'})
        i = 0
        lgnd = []

        for sensor in lista:
            ax.plot(mtu[sensor], color = '0.8', label = "_nolegend_")

        for sensor in lista:
            v_moving = mtu[sensor].rolling(window = movil).mean()
            ax.plot(v_moving, color = colors[i])
            i +=1
            lgnd.append(sensor)

        ax.set(xlabel = "Tiempo [min]", ylabel = "Señal Adquirida [p.u.]", xlim = [0, 1440], ylim = [0, 1])
        plt.legend(lgnd, loc = 'upper right', bbox_to_anchor=(1.13,1),borderaxespad=0)
        plt.grid()

        day_date = date(year, month,day)
        # Save images
        plt.savefig(folder_images  + tab  + folder_images_child + tab + f"Grupo_{grupos[j]}" + ".png")
        plt.close()
        j += 1
```

```python
# Complete
# Select day of analysis
month = 12
year = 2021
days = [str(i).zfill(2) for i in range(1, monthrange(year, month)[1] + 1)]
month_date = date(year,month, 1)
rows = ['MEDIA [mV]', 'STD [%]', 'MIN [mV]','MAX [mV]', 'MEDIANA [mV]']
if machine == "windows":
        folder_images = mtu_per_days + month_date.strftime("%b_%Y").upper()

if not os.path.exists(folder_images):
        os.makedirs(folder_images)

for day in days:

    day = int(day)
    day_date = date(year,month, day)

    # Load mtu file

    dir_mtu =  mtu_file_name + day_date.strftime("%b%d_%Y").upper() + ".txt"
    mtu = pd.read_csv(dir_mtu, delimiter = "\t")

    # Normalize data
    sensors = ["P" + str(i) for i in range(1, 73)]
    sensors0 = ["P" + str(i) for i in[15, 16, 17, 19, 21, 22, 23, 50, 51, 52, 58, 59, 60]]

    for sensor in sensors0:
        sensors.remove(sensor)

    # Here goes the plot
    mydpi = 200
    ancho = 1920
    ratio = 10/4
    movil = 30
    colors = ['#FF0000','#FF4000','#FF8000','#FFC000','#ffff00','#bfff00','#00ff00','#00ff80','#00ffff','#00bfff','#007fff','#0000ff','#8000ff','#c000ff','#ff00ff','#ff007f']
    listas = [sensors[i:i + 16] for i in range(0, len(sensors), 16)]

    if machine == "windows":
        folder_images_child = day_date.strftime("%b%d_%Y").upper()

    if not os.path.exists(folder_images + tab + folder_images_child):
        os.makedirs(folder_images + tab + folder_images_child)

    mtu = mtu[sensors]
    mtu_report = mtu.copy()
    mtu_report = mtu_report.replace(0, np.NaN)
    df_relacion = pd.DataFrame(columns= sensors, index=rows)

    # Normalize sensors
    for sensor in sensors:
        max = mtu[sensor].max()

        if max != 0:
            mtu[sensor] = mtu[sensor]/max
        else:
            mtu[sensor] = 0

        df_relacion[sensor][rows[0]] = mtu_report[sensor].mean()
        if mtu_report[sensor].mean() == 0:
            df_relacion[sensor][rows[1]] = 0
        else:
            df_relacion[sensor][rows[1]] = 100*(mtu_report[sensor].std() / mtu_report[sensor].mean())

        df_relacion[sensor][rows[2]] = (mtu_report[sensor]).min()
        df_relacion[sensor][rows[3]] = (mtu_report[sensor]).max()
        df_relacion[sensor][rows[4]] = (mtu_report[sensor]).median()

    for sensor in sensors:
        for i in range(df_relacion.shape[0]):
            df_relacion[sensor][i] = round(df_relacion[sensor][i],2)
    df_relacion_transposed = df_relacion.T
    df_relacion_transposed.to_excel(folder_images  + tab  + folder_images_child + tab + "tabla.xlsx")
```
